# NBA Team Salaries

## Data Processing

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 100)
```

```python
# Get 2001 - 2021 data

df_2001 = pd.read_html('https://hoopshype.com/salaries/2000-2001/', index_col=0)[0]
df_2002 = pd.read_html('https://hoopshype.com/salaries/2001-2002/', index_col=0)[0]
df_2003 = pd.read_html('https://hoopshype.com/salaries/2002-2003/', index_col=0)[0]
df_2004 = pd.read_html('https://hoopshype.com/salaries/2003-2004/', index_col=0)[0]
df_2005 = pd.read_html('https://hoopshype.com/salaries/2004-2005/', index_col=0)[0]
df_2006 = pd.read_html('https://hoopshype.com/salaries/2005-2006/', index_col=0)[0]
df_2007 = pd.read_html('https://hoopshype.com/salaries/2006-2007/', index_col=0)[0]
df_2008 = pd.read_html('https://hoopshype.com/salaries/2007-2008/', index_col=0)[0]
df_2009 = pd.read_html('https://hoopshype.com/salaries/2008-2009/', index_col=0)[0]
df_2010 = pd.read_html('https://hoopshype.com/salaries/2009-2010/', index_col=0)[0]

df_2011 = pd.read_html('https://hoopshype.com/salaries/2010-2011/', index_col=0)[0]
df_2012 = pd.read_html('https://hoopshype.com/salaries/2011-2012/', index_col=0)[0]
df_2013 = pd.read_html('https://hoopshype.com/salaries/2012-2013/', index_col=0)[0]
df_2014 = pd.read_html('https://hoopshype.com/salaries/2013-2014/', index_col=0)[0]
df_2015 = pd.read_html('https://hoopshype.com/salaries/2014-2015/', index_col=0)[0]
df_2016 = pd.read_html('https://hoopshype.com/salaries/2015-2016/', index_col=0)[0]
df_2017 = pd.read_html('https://hoopshype.com/salaries/2016-2017/', index_col=0)[0]
df_2018 = pd.read_html('https://hoopshype.com/salaries/2017-2018/', index_col=0)[0]
df_2019 = pd.read_html('https://hoopshype.com/salaries/2018-2019/', index_col=0)[0]
df_2020 = pd.read_html('https://hoopshype.com/salaries/2019-2020/', index_col=0)[0]

df_2021 = pd.read_html('https://hoopshype.com/salaries/2020-2021/', index_col=0)[0]

df_2011.head()
```

```python
# Merge all data into a single DataFrame

merge1 = pd.merge(df_2001, df_2002, on='Team', how='outer')
merge2 = pd.merge(merge1, df_2003, on='Team', how='outer')
merge3 = pd.merge(merge2, df_2004, on='Team', how='outer')
merge4 = pd.merge(merge3, df_2005, on='Team', how='outer')
merge5 = pd.merge(merge4, df_2006, on='Team', how='outer')
merge6 = pd.merge(merge5, df_2007, on='Team', how='outer')
merge7 = pd.merge(merge6, df_2008, on='Team', how='outer')
merge8 = pd.merge(merge7, df_2009, on='Team', how='outer')
merge9 = pd.merge(merge8, df_2010, on='Team', how='outer')
merge10 = pd.merge(merge9, df_2011, on='Team', how='outer')

merge11 = pd.merge(merge10, df_2012, on='Team', how='outer')
merge12 = pd.merge(merge11, df_2013, on='Team', how='outer')
merge13 = pd.merge(merge12, df_2014, on='Team', how='outer')
merge14 = pd.merge(merge13, df_2015, on='Team', how='outer')
merge15 = pd.merge(merge14, df_2016, on='Team', how='outer')
merge16 = pd.merge(merge15, df_2017, on='Team', how='outer')
merge17 = pd.merge(merge16, df_2018, on='Team', how='outer')
merge18 = pd.merge(merge17, df_2019, on='Team', how='outer')
merge19 = pd.merge(merge18, df_2020, on='Team', how='outer')
merge20 = pd.merge(merge19, df_2021, on='Team', how='outer')

merge20.head()
```

```python
# Keep the following columns

keep = ['Team', 
        '2000/01', '2001/02', '2002/03', '2003/04', '2004/05', '2005/06', '2006/07', '2007/08', '2009/10', '2010/11', 
        '2011/12', '2012/13', '2013/14', '2014/15', '2015/16', '2016/17', '2017/18', '2019/20', '2020/21']

df = merge20[keep].copy()
df.head()
```

```python
# Convert salary to numeric

for c in keep[1:]:
    df[c] = pd.to_numeric(df[c].str[1:].replace(',', '', regex=True))

df.head()
```

```python
# Define team dictionary for renaming
teams_dict = {
    'Boston': 'Boston Celtics', 
    'Brooklyn': 'Brooklyn Nets', 
    'New York': 'New York Knicks', 
    'Philadelphia': 'Philadelphia 76ers', 
    'Toronto': 'Toronto Raptors', 
    'Chicago': 'Chicago Bulls', 
    'Cleveland': 'Cleveland Cavaliers', 
    'Detroit': 'Detroit Pistons', 
    'Indiana': 'Indiana Pacers', 
    'Milwaukee': 'Milwaukee Bucks', 
    'Atlanta': 'Atlanta Hawks', 
    'Charlotte': 'Charlotte Bobcats', 
    'Miami': 'Miami Heat', 
    'Orlando': 'Orlando Magic', 
    'Washington': 'Washington Wizards', 
    'Denver': 'Denver Nuggets', 
    'Minnesota': 'Minnesota Timberwolves', 
    'Oklahoma City': 'Oklahoma City Thunder', 
    'Portland': 'Portland Trail Blazers', 
    'Utah': 'Utah Jazz', 
    'Golden State': 'Golden State Warriors', 
    'LA Clippers': 'Los Angeles Clippers', 
    'LA Lakers': 'Los Angeles Lakers', 
    'Phoenix': 'Phoenix Suns', 
    'Sacramento': 'Sacramento Kings', 
    'Dallas': 'Dallas Mavericks', 
    'Houston': 'Houston Rockets', 
    'Memphis': 'Memphis Grizzlies', 
    'New Orleans': 'New Orleans Hornets', 
    'San Antonio': 'San Antonio Spurs', 
}

# Define division dict
divisions_dict = {
    'Boston Celtics': 'Atlantic', 
    'Brooklyn Nets': 'Atlantic', 
    'New York Knicks': 'Atlantic', 
    'Philadelphia 76ers': 'Atlantic', 
    'Toronto Raptors': 'Atlantic', 
    'Chicago Bulls': 'Central', 
    'Cleveland Cavaliers': 'Central', 
    'Detroit Pistons': 'Central', 
    'Indiana Pacers': 'Central', 
    'Milwaukee Bucks': 'Central', 
    'Atlanta Hawks': 'Southeast', 
    'Charlotte Bobcats': 'Southeast', 
    'Miami Heat': 'Southeast', 
    'Orlando Magic': 'Southeast', 
    'Washington Wizards': 'Southeast', 
    'Denver Nuggets': 'Northwest', 
    'Minnesota Timberwolves': 'Northwest', 
    'Oklahoma City Thunder': 'Northwest', 
    'Portland Trail Blazers': 'Northwest', 
    'Utah Jazz': 'Northwest', 
    'Golden State Warriors': 'Pacific', 
    'Los Angeles Clippers': 'Pacific', 
    'Los Angeles Lakers': 'Pacific', 
    'Phoenix Suns': 'Pacific', 
    'Sacramento Kings': 'Pacific', 
    'Dallas Mavericks': 'Southwest', 
    'Houston Rockets': 'Southwest', 
    'Memphis Grizzlies': 'Southwest', 
    'New Orleans Hornets': 'Southwest', 
    'San Antonio Spurs': 'Southwest', 
}

# Define conferences dict
conferences_dict = {
    'Atlantic': 'Eastern', 
    'Central': 'Eastern', 
    'Southeast': 'Eastern', 
    'Northwest': 'Western', 
    'Pacific': 'Western', 
    'Southwest': 'Western'
}

# Change names of teams
df = df.replace(teams_dict)

# Add divisions
df['Division'] = df['Team'].map(divisions_dict)

# Add conferences
df['Conference'] = df['Division'].map(conferences_dict)

# Sort according to divisions, then teams
df = df.sort_values(['Conference', 'Division', 'Team']).reset_index(drop=True)

df.head()
```

```python
# Transpose the dataframe

df = df[keep].T
df = df.rename(columns=df.loc['Team'])
df = df[df.index != 'Team']
df.head()
```

```python
# Compute average salaries across teams

teams = df.columns[:-1]
df['avg'] = df[teams].apply(lambda x: x.mean(), axis=1)
df.head()
```

## Data Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Set font to Serif
plt.rcParams['font.family'] = 'Serif'
```

```python
# Create 30 subplots, 6 rows, 5 cols
fig, axs = plt.subplots(6, 5, figsize=(18, 18), facecolor='white')

# Adjust space between subplots
fig.subplots_adjust(wspace=.1, hspace=.4)

# Target team
my_team = 'Golden State Warriors'

# Loop through all axes and plot
for ax, current_team in zip(axs.ravel(), df.columns):
    
    # Loop through all teams
    for team in df.columns:
        
        # Set characteristics of lines according to team
        if team == my_team:
            color = 'brown'
            linewidth = 1.5
            zorder = 2
            label = my_team
        elif team == current_team != my_team:
            color = 'steelblue'
            linewidth = 1.5
            zorder = 2
            label = 'Specific Team'
        elif team == 'avg':
            color = 'slategrey'
            linewidth = 1.5
            zorder = 2
            label = 'Average'
        else:
            color = 'lightgrey'
            linewidth = .3
            zorder = 1
            label = None
        
        # Plot line graph
        ax.plot(df.index, df[team], color=color, linewidth=linewidth, zorder=zorder, label=label)
    
    # Remove borders
        for sp in ax.spines.values():
            sp.set_visible(False)
            
    # Remove ticks 
    ax.tick_params(axis='both', which='both', left=False, bottom=False, labelleft=False, labelbottom=False)
    
    # Set title of subplot
    ax.set_title(current_team)

# Get labels of first subplot
lines, labels = axs[0][0].get_legend_handles_labels()
# Add legend
fig.legend(lines, labels, loc='center', bbox_to_anchor=(0.5, 0.1), ncol=3, frameon=False, fontsize=14)

# Add title for entire graph
fig.suptitle('Salaries of NBA Teams (2001-2021)', y=0.94, fontsize=20)

# Add text for data source reference
fig.text(.5, .06, 'Created by H | Data: HoopsHype', horizontalalignment='center', fontsize=14)
```
