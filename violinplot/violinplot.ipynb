```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_palette('Paired')
sns.set_style("whitegrid")
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
with sns.axes_style('white'):
    fig, axes = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(10, 10))
    sns.boxplot(data=df, y='age', ax=axes[0]).set(xticks=[], yticks=[], ylabel='')
    sns.violinplot(data=df, y='age', ax=axes[2]).set(xticks=[], yticks=[], ylabel='')
    sns.despine(bottom=True, left=True)
```

```python
plt.figure(figsize=(5, 10))
sns.violinplot(data=df, y='age').set(ylabel='Age')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='age').set(xlabel='Age')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='class', y='age').set(xlabel='Class', ylabel='Age')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='class', y='age', hue='sex').set(xlabel='Class', ylabel='Age')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='class', y='age', hue='sex', split=True).set(xlabel='Class', ylabel='Age')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='class', y='age', hue='sex', split=True, inner='quartile')\
    .set(xlabel='Class', ylabel='Age')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='class', y='age', hue='sex', split=True, inner='quartile', scale='count')\
    .set(xlabel='Class', ylabel='Age')
```
